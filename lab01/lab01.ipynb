```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(2137)
```

```python
mu = 0
sigma = 1
m = 10000

x = None
y = None
x = np.sort(np.random.normal(mu, sigma, m))
y = np.zeros(m)
k = int(np.sqrt(m))


plt.scatter(x, y, s=200, marker="|")
count, bins, ignored = plt.hist(x, k, density=True)
temp=np.linspace(-10.0, 10.0, num=1000)
plt.plot(temp, 1/(sigma * np.sqrt(2 * np.pi)) *np.exp( - (temp - mu)**2 / (2 * sigma**2) ), linewidth=2, color='r')
plt.vlines(x=x, ymin=-0.01, ymax=0.01, colors='purple', ls='--', lw=1, label='vline_multiple - full height')
plt.title("Rozk≈Çad normalny")
plt.xlabel("x")
plt.ylabel("y")
plt.xlim([-10, 10])
plt.ylim([0, 0.4])
plt.grid()

plt.show()
```

```python
for m in [100,1000,10000,100000]:
    x = None
    y = None
    mu1 = 0
    sigma1 = 1
    mu2 = 3
    sigma2 = 1
    mu3 = -5
    sigma3 = 1
    x1 = np.sort(np.random.normal(mu1, sigma1, int(m*0.5)))
    x2 = np.sort(np.random.normal(mu2, sigma2, int(m*0.3)))
    x3 = np.sort(np.random.normal(mu3, sigma3, int(m*0.2)))
    x = np.sort(np.concatenate((x1,x2,x3)))
    y = np.zeros(m)
    k = int(np.sqrt(m))

    wsp1=0.5
    wsp2=0.3
    wsp3=0.2


    plt.scatter(x, y, s=200, marker="|")
    count, bins, ignored = plt.hist(x, k, density=True)
    temp=np.linspace(-10.0, 10.0, num=1000)
    plt.plot(temp, wsp1/(sigma1 * np.sqrt(2 * np.pi)) *np.exp( - (temp - mu1)**2 / (2 * sigma1**2)) + wsp2/(sigma2 * np.sqrt(2 * np.pi)) *np.exp( - (temp - mu2)**2 / (2 * sigma2**2)) + wsp3/(sigma3 * np.sqrt(2 * np.pi)) *np.exp( - (temp - mu3)**2 / (2 * sigma3**2)), linewidth=2, color='r')
    plt.vlines(x=x, ymin=-0.01, ymax=0.01, colors='purple', ls='--', lw=1, label='vline_multiple - full height')
    plt.title("Rozk≈Çad normalny")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.xlim([-10, 10])
    plt.ylim([0, 0.4])
    plt.grid()

    plt.show()
```
