```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
```

```python
m = 1000
sigma = 1 # h
w1 = .6
w2 = .4
mu1 = -7
mu2 = [-7, -4, 3]
r = 0.2

x = np.linspace(-10.0, 10.0, num=100)
fix, ax = plt.subplots(6, 3, figsize=(36, 24))

for pj, mu2 in enumerate(mu2):
    k = [1, 2, 5, 10, 20, 500]
    for pi, k in enumerate(k):
        x1 = np.random.normal(mu1, 1, int(m * w1))
        x2 = np.random.normal(mu2, 1, int(m * w2))
        xi = np.sort(np .concatenate((x1, x2)))
        y = w1/(np.sqrt(2 * np.pi)) * np.exp(- ( x - mu1)**2 / (2 * sigma**2)) + w2/(np.sqrt(2 * np.pi)) * np.exp(- ( x - mu2)**2 / (2 * sigma**2))

        distance = np.zeros(m)
        for i in range(m):
            if i < k:
                distance[i] = np.absolute(xi[i+k] - xi[i])
                continue
            if i > m-k-1:
                distance[i] = np.absolute(xi[i] - xi[i-k])
                continue
            distance[i] = np.minimum(np.absolute(xi[i]-xi[i-k]), np.absolute(xi[i]-xi[i+k]) )   
        
    
        distance_sorted = np.sort(distance)
        distance_to_catch = distance_sorted[int(m*(1 - r))]
        elements_x = []
        for i in range(m-1):
            if distance[i] >= distance_to_catch:
                elements_x.append(xi[i])

        ax[pi, pj].plot(x, y , linewidth=2, color="r")
        ax[pi, pj].set_xlim([-10, 10])
        ax[pi, pj].set_ylim([0, .5])
        ax[pi, pj].grid()
        ax[pi, pj].vlines(x=xi, ymin=0, ymax=.02, colors='purple', ls='-')
        ax[pi, pj].vlines(x=elements_x, ymin=.03, ymax=.04, colors='red', ls='-')
        title = ''
        if pi == 0:
            title = '60% N(' + str(mu1) + ', 1) + 40% N(' + str(mu2) + ', 1)'+ ', r=' + str(r) + '\n'
        ax[pi, pj].set_title(title + 'k=' + str(k))
plt.show()
```
